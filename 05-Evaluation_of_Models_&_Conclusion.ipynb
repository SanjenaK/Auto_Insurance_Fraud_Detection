The Decision Tree and Random Forest models also obtained higher precisions, F1 scores, and AUCs compared to the Logistic Regression, Naive Bayes, KNN, SVM, amd MLP classifiers applied.  Although SVM and MLP often show good accuracies, these models were outperformed by the Decision Tree and Random Forest algorithms in this project.  Overall, training classifiers with RFE showed better performance than without using RFE or PCA and training models with PCA decreased the classification algorithms' performance except for Naive Bayes and KNN.  The independent features in the training set was scaled and an oversampling technique using SMOTE was applied to the minority class.  The F1 scores being <0.62 suggests that using another technique to balance the dataset could help improve the classification models.
